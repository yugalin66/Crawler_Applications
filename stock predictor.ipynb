```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
import joblib
import openpyxl as xl
from openpyxl import Workbook

data = pd.read_csv('stock.csv')
X = data.drop(columns=["no", "today close", "change ratio", "profitable"])
y = data['profitable']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)
model = DecisionTreeClassifier()
model.fit(X_train, y_train)
predictions = model.predict(X_test)
score = accuracy_score(y_test, predictions)
score
#joblib.dump(model, "StockPredictor.joblib")

# model = joblib.load("StockPredictor.joblib")
# wb = xl.load_workbook("stock.xlsx")
# sheet = wb.worksheets[0]
# prediction = []
# for row in range(40, 50):
#     no = sheet.cell(row, 1).value
#     week = sheet.cell(row, 2).value
# #     if week[0] == "-":
# #         week = week.replace("-", "")
# #         week = 0 - int(week)
#     last = sheet.cell(row, 3).value
#     today = sheet.cell(row, 4).value
#     prediction = model.predict([[week,last,today]])
#     if prediction:
#         print(no)
```
