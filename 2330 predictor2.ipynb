```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
import joblib
import openpyxl as xl
from openpyxl import Workbook


############### can be modified ##################

stock_number = "2330"

################################## get today's data #############################################

wb = xl.load_workbook(stock_number + ".xlsx")
today_sheet = wb["today_data"]
today_data = []
price = []
for j in range(2, today_sheet.max_column + 1):
    price.append(today_sheet.cell(2, j).value)
today_data.append(price)

################################## transfer .xlsx to .csv file ######################################

file = pd.read_excel(stock_number + ".xlsx")
file.to_csv(stock_number + ".csv")
data = pd.read_csv(stock_number + ".csv")

# ######################################### machine learning ##########################################

X = data.drop(columns=["date", "Unnamed: 0", "prediction"])
y = data['prediction']

today_predict = []
for i in range(1000):
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)
    model = DecisionTreeClassifier()
    model.fit(X_train.values, y_train.values)
    #predictions = model.predict(X_test)
    today_predict.append(model.predict(today_data))
    
today_predict.sort()
predict_unique = []
probs = []
for predict in today_predict:
    if predict not in predict_unique:
        predict_unique.append(predict)
for unique in predict_unique:
    probs.append(today_predict.count(unique))
for i in range(len(probs)):
    print(f"{predict_unique[i]} {probs[i]/10}%")

# score = accuracy_score(y_test, predictions)
# score
#joblib.dump(model, "StockPredictor.joblib")

# model = joblib.load("StockPredictor.joblib")
# wb = xl.load_workbook("stock.xlsx")

```
